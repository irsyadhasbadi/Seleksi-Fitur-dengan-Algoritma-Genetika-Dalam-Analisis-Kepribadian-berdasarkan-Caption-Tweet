```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Akurasi per label
akurasi_pop10 = {'IE': 0.63, 'NS': 0.54, 'TF': 0.57, 'JP': 0.55}
akurasi_pop15 = {'IE': 0.61, 'NS': 0.54, 'TF': 0.58, 'JP': 0.55}

# Precision per kelas
precision_pop10 = {'E': 0.68, 'I': 0.37, 'N': 0.60, 'S': 0.44, 'F': 0.61, 'T': 0.41, 'J': 0.70, 'P': 0.53}
precision_pop15 = {'E': 0.68, 'I': 0.33, 'N': 0.61, 'S': 0.45, 'F': 0.62, 'T': 0.43, 'J': 0.61, 'P': 0.54}

# Recall per kelas
recall_pop10 = {'E': 0.84, 'I': 0.20, 'N': 0.66, 'S': 0.39, 'F': 0.83, 'T': 0.19, 'J': 0.18, 'P': 0.92}
recall_pop15 = {'E': 0.82, 'I': 0.19, 'N': 0.63, 'S': 0.42, 'F': 0.83, 'T': 0.19, 'J': 0.34, 'P': 0.78}

# F1-score per kelas
f1_pop10 = {'E': 0.75, 'I': 0.26, 'N': 0.63, 'S': 0.42, 'F': 0.70, 'T': 0.26, 'J': 0.28, 'P': 0.67}
f1_pop15 = {'E': 0.74, 'I': 0.24, 'N': 0.62, 'S': 0.43, 'F': 0.71, 'T': 0.26, 'J': 0.43, 'P': 0.64}

# Fungsi plot grouped bar
def plot_grouped(data1, data2, title, ylabel, labels):
    x = np.arange(len(labels))
    width = 0.35

    fig, ax = plt.subplots(figsize=(10,5))
    bars1 = ax.bar(x - width/2, [data1[l] for l in labels], width, label='Populasi 10')
    bars2 = ax.bar(x + width/2, [data2[l] for l in labels], width, label='Populasi 15')

    for bar in bars1 + bars2:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
                f"{bar.get_height():.2f}", ha='center', fontsize=8)

    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylim(0, 1)
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Plot akurasi per label
plot_grouped(akurasi_pop10, akurasi_pop15, "Perbandingan Akurasi per Label", "Akurasi", list(akurasi_pop10.keys()))

# Plot precision per kelas
plot_grouped(precision_pop10, precision_pop15, "Perbandingan Precision per Kelas", "Precision", list(precision_pop10.keys()))

# Plot recall per kelas
plot_grouped(recall_pop10, recall_pop15, "Perbandingan Recall per Kelas", "Recall", list(recall_pop10.keys()))

# Plot f1-score per kelas
plot_grouped(f1_pop10, f1_pop15, "Perbandingan F1-score per Kelas", "F1-score", list(f1_pop10.keys()))
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Data Akurasi per label
labels = ['IE', 'NS', 'TF', 'JP']
akurasi_10 = [0.63, 0.54, 0.57, 0.55]
akurasi_15 = [0.61, 0.54, 0.58, 0.56]
akurasi_20 = [0.65, 0.57, 0.58, 0.54]

# Precision per kelas (E, I, N, S, F, T, J, P)
precision_10 = [0.68, 0.37, 0.60, 0.44, 0.61, 0.41, 0.70, 0.53]
precision_15 = [0.68, 0.33, 0.61, 0.45, 0.62, 0.43, 0.61, 0.54]
precision_20 = [0.68, 0.37, 0.60, 0.47, 0.61, 0.40, 0.57, 0.53]

# Recall per kelas
recall_10 = [0.84, 0.20, 0.66, 0.39, 0.83, 0.19, 0.18, 0.92]
recall_15 = [0.82, 0.19, 0.63, 0.42, 0.83, 0.19, 0.34, 0.78]
recall_20 = [0.92, 0.09, 0.80, 0.24, 0.86, 0.15, 0.33, 0.75]

# F1-score per kelas
f1_10 = [0.75, 0.26, 0.63, 0.42, 0.70, 0.26, 0.28, 0.67]
f1_15 = [0.74, 0.24, 0.62, 0.43, 0.71, 0.26, 0.43, 0.64]
f1_20 = [0.78, 0.15, 0.69, 0.32, 0.71, 0.22, 0.42, 0.62]

kelas = ['E', 'I', 'N', 'S', 'F', 'T', 'J', 'P']

x = np.arange(len(labels))
x_kelas = np.arange(len(kelas))
width = 0.25

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Perbandingan Hasil Evaluasi (Populasi 10, 15, 20)', fontsize=16)

# 1. Akurasi
ax = axes[0, 0]
rects1 = ax.bar(x - width, akurasi_10, width, label='Pop 10')
rects2 = ax.bar(x, akurasi_15, width, label='Pop 15')
rects3 = ax.bar(x + width, akurasi_20, width, label='Pop 20')
ax.set_ylabel('Akurasi')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 1)
ax.legend()
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=8)

# 2. Precision
ax = axes[0, 1]
rects1 = ax.bar(x_kelas - width, precision_10, width, label='Pop 10')
rects2 = ax.bar(x_kelas, precision_15, width, label='Pop 15')
rects3 = ax.bar(x_kelas + width, precision_20, width, label='Pop 20')
ax.set_ylabel('Precision')
ax.set_xticks(x_kelas)
ax.set_xticklabels(kelas)
ax.set_ylim(0, 1)
ax.legend()
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=8)

# 3. Recall
ax = axes[1, 0]
rects1 = ax.bar(x_kelas - width, recall_10, width, label='Pop 10')
rects2 = ax.bar(x_kelas, recall_15, width, label='Pop 15')
rects3 = ax.bar(x_kelas + width, recall_20, width, label='Pop 20')
ax.set_ylabel('Recall')
ax.set_xticks(x_kelas)
ax.set_xticklabels(kelas)
ax.set_ylim(0, 1)
ax.legend()
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=8)

# 4. F1-score
ax = axes[1, 1]
rects1 = ax.bar(x_kelas - width, f1_10, width, label='Pop 10')
rects2 = ax.bar(x_kelas, f1_15, width, label='Pop 15')
rects3 = ax.bar(x_kelas + width, f1_20, width, label='Pop 20')
ax.set_ylabel('F1-score')
ax.set_xticks(x_kelas)
ax.set_xticklabels(kelas)
ax.set_ylim(0, 1)
ax.legend()
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=8)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
